### Simple Machine Learning Model

```python
## Setting Independent and Dependent features
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
y
```

```python
## Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=2023)
X_train
```

### Preprocessing

```python
## Standarizing the dataset
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
X_train
```

### Training the model

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
# Printing the coefficients and the intercept
regression.coef_
```

```python
regression.intercept_
```

```python
## On which parmaeters the model has been trained?
regression.get_params()
```

```python
## Predicting with the test data
reg_pred = regression.predict(X_test)
reg_pred
```

## Assumptions

```python
## Scatter plot for the prediction 
plt.scatter(y_test,reg_pred)
```

```python
## Residuals
residuals = y_test - reg_pred
## Plotting Residuals
sns.displot(residuals,kind='kde')
```

```python
## Scatter plot with respect to prediction and residuals
plt.scatter(reg_pred, residuals)
```

We have a unifrom distribution, so we can assume the model is performing well.

### Performance of the model

MAE and MSE

```python
MAE = mean_absolute_error(y_test,reg_pred)
MSE = mean_squared_error(y_test,reg_pred)

print('The mean absolute error MAE is:', MAE)
print('The mean absolute error MAE is:', MSE)
```

***R square and Adjusted R square***

R-square: R^2 = 1 - SSR/SST

Where: SSR (sum of squares of residuals), SST (total sum of squares)

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print('The R square is:', score)
```

Adjusted R2 = 1 - [(1-R2)*(n-1)/(n-k-1)]

Where: R2(score), n(number of observations), k(number of predictors).

```python
adj_r2 = 1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print('The adjusted R-squared is: ', adj_r2)
```
